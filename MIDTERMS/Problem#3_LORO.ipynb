```python
import control
import matplotlib.pyplot as plt
import numpy as np

# 1. Define Transfer Function G(s) = (4s^2 + 2s + 6) / (16s^4 + 24s^3 + 52s^2 + 36s)
num = [4, 2, 6]
den = [16, 24, 52, 36, 0]  # Note the 0 for the constant term

sys = control.TransferFunction(num, den)

print(f"Transfer Function G(s):\n{sys}")
```

```python
# 2. Plot Step Response
t, y = control.step_response(sys)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(t, y, color='#0ea5e9', linewidth=2)
plt.title('Step Response', fontsize=12, fontweight='bold')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True, linestyle='--', alpha=0.7)
```

```python
# 3. Generate Pole-Zero Map
plt.subplot(1, 2, 2)
control.pzmap(sys, plot=True)
plt.title('Pole-Zero Map', fontsize=12, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
```

```python
plt.tight_layout()
plt.savefig('system_analysis.png')
plt.show()
```
