```python
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt

num = [3]
den = [8, 12, 26, 18, 0]

B = ctrl.TransferFunction(num, den)

print("Transfer Function B(s):")
print(B)
```

```python
poles = ctrl.poles(B)
zeros = ctrl.zeros(B)
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
plt.figure(figsize=(10, 6), dpi=100)
t, y = ctrl.step_response(B)
plt.plot(t, y, color='#0D9488', linewidth=2)
plt.title('System Step Response Analysis', fontsize=14, fontweight='bold')
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('Amplitude X3(t)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6), dpi=100)
ctrl.pzmap(B, grid=True)
plt.title('Pole-Zero Map of B(s)', fontsize=14, fontweight='bold')
plt.show()
```
