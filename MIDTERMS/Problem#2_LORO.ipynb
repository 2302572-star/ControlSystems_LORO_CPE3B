```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# 1. Define the transfer function coefficients
# G(s) = 1 / (s^4 + 2s^3 + 3s^2 + 2s)
num = [1]
den = [1, 2, 3, 2, 0]

# 2. Create the system object
sys = signal.TransferFunction(num, den)

# 3. Calculate unit step response
t, y = signal.step(sys)

# 4. Calculate poles
poles = sys.poles

# 5. Setup the plotting area
plt.figure(figsize=(12, 5))
```

```python
# --- Step Response ---
plt.subplot(1, 2, 1)
plt.plot(t, y)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output Theta_2')
plt.grid(True)
```

```python
# --- Pole-Zero Map ---
plt.subplot(1, 2, 2)
plt.scatter(np.real(poles), np.imag(poles), marker='x', color='red', label='Poles')
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title('Pole-Zero Map')
plt.xlabel('Real Axis')
plt.ylabel('Imaginary Axis')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig('system_analysis.png')
plt.show()
```
