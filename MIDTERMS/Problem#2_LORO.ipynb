```python
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt

num = [1]
den = [1, 2, 3, 2, 0]

B = ctrl.TransferFunction(num, den)
print("Transfer Function:")
print(f"B(s) = {B}")
```

```python
poles = ctrl.poles(B)
zeros = ctrl.zeros(B)
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
t, y = ctrl.step_response(B)

plt.figure(figsize=(10, 5))
plt.plot(t, y, color='#0d9488', linewidth=2)
plt.title('System Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
ctrl.pzmap(B, grid=True)
plt.title('Pole-Zero Map of the System')
plt.show()
```
