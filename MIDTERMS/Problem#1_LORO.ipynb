```python
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt

num = [1, 1]
den = [1, 4, 7, 6, 3]

B = ctrl.TransferFunction(num, den)

print("Transfer Function:")
print(f"B(s) = {B}")
```

```python
poles = ctrl.poles(B)
zeros = ctrl.zeros(B)

print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
t, y = ctrl.step_response(B)

plt.figure(figsize=(10, 5))
plt.plot(t, y, color='#0D9488', linewidth=2)
plt.title('Step Response of the System', fontsize=14, fontweight='bold')
plt.xlabel('Time (s)')
plt.ylabel('Output')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
ctrl.pzmap(B, grid=True)
plt.title('Pole-Zero Map', fontsize=14, fontweight='bold')
plt.show()
```
