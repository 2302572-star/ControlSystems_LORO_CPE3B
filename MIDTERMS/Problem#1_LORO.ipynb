```python
import control as ctrl
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the Transfer Function H(s) = (s + 1) / (s^4 + 4s^3 + 7s^2 + 6s + 3)
num = [1, 1]
den = [1, 4, 7, 6, 3]
sys = ctrl.TransferFunction(num, den)

print(f"System Transfer Function:\n{sys}")
```

```python
t, y = ctrl.step_response(sys)

plt.figure(figsize=(12, 5))

# Plot Step Response
plt.subplot(1, 2, 1)
plt.plot(t, y, 'b', linewidth=2)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
```

```python
# 3. Pole-Zero Map
plt.subplot(1, 2, 2)
ctrl.pole_zero_plot(sys)
plt.title('Pole-Zero Map')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Display Poles and Zeros in console
print(f"Poles: {ctrl.poles(sys)}")
print(f"Zeros: {ctrl.zeros(sys)}")
```
